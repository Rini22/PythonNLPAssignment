```python
from weather import Weather, Unit


df.reset_index(drop=True).head(10)
```

```python
#populating column weather using yahoo weather api and showing it for 20 cities as it's time extensive task
def valuation_formula(x):
    weather = Weather(unit=Unit.CELSIUS)
    loc=weather.lookup_by_location(x).condition.text
    return loc
dfNew=df.head(19)
dfNew["weather"]=dfNew.head(19).apply(lambda row:valuation_formula(row["city"]),axis=1)
```
